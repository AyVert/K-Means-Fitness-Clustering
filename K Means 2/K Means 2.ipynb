```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data generation
# Create 2 arrays of 10 points each, one centered at (1,1) and the other at (5,5)
# The points are normally distributed around the center with a standard deviation of 1
# The first column is the x coordinate and the second column is the y coordinate

def generate_data():
	# Create the first array
	x1 = np.random.randn(200,2) + np.array([1,1])
	# Create the second array
	x2 = np.random.randn(200,2) + np.array([3,3])
	# Concatenate the two arrays
	X = np.concatenate((x1,x2), axis = 0)
	return X
```

```python
# Create 2 random points as the initial centroids that are within the range of the data
def initialize_centroids(X, K):
	# Randomly pick K data points as the initial centroids
	# np.random.permutation(X.shape[0]) returns a random permutation of integers from 0 to X.shape[0]
	# [:K] returns the first K elements of the array
	centroids = X[np.random.permutation(X.shape[0])[:K]]
	return centroids
```

```python
# Assign each data point to the closest centroid
def assign_centroids(X, centroids):
	# Create an array that will hold the index of the closest centroid for each data point
	# The array will have the same number of rows as X and 1 column
	# The array will be initialized with zeros
	# The data type of the array will be integer
	# The array will be of size 20 x 1
	# The array will be initialized with zeros
	# The data type of the array will be integer
	# The array will be of size 20 x 1
	idx = np.zeros(X.shape[0], dtype = int)
	# Loop through each data point
	for i in range(X.shape[0]):
		# Initialize the minimum distance to a large number
		min_dist = 1000000
		# Loop through each centroid
		for j in range(centroids.shape[0]):
			# Compute the distance between the data point and the centroid
			dist = np.sum((X[i,:] - centroids[j,:]) ** 2)
			# If the distance is less than the minimum distance
			if dist < min_dist:
				# Update the minimum distance
				min_dist = dist
				# Assign the index of the centroid to the data point
				idx[i] = j
	return idx
```

```python
#idx = assign_centroids(X, centroids)


def plot_data(X, idx, centroids):
	# Plot the data and the centroids with the data points colored according to the centroid index
	plt.scatter(X[:,0], X[:,1], c = idx)
	# Plot the centroids in star shape
	plt.scatter(centroids[:,0], centroids[:,1], marker = '*', s = 150)
	plt.xlabel('Average running speed')
	plt.ylabel('Average Heart Rate')
	plt.show()

#plot_data(X, idx, centroids)
```

```python
# Update the centroids
def update_centroids(X, K, idx):
	# Create an array that will hold the centroids
	centroids = np.zeros((K, X.shape[1]))
	# Loop through each centroid
	for i in range(K):
		# Assign the data points that belong to the centroid
		centroid_data = X[idx == i,:]
		# Compute the mean of the data points
		centroids[i,:] = np.mean(centroid_data, axis = 0)
	return centroids
```

```python
X = generate_data()
K = 3
```

```python
centroids = initialize_centroids(X, K)
idx = assign_centroids(X, centroids)
plot_data(X, idx, centroids)
```
